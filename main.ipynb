```python
import scipy.signal
def gaborfilter(I, S, F, W, P):
    size = int(1.5 / S)
    F = (S ** 2) / np.sqrt(2 * np.pi)
    k = 1
    G = [[0] * (2*size) for _ in range(2*size)]
    for x in range(-size, size):
        for y in range(-size, size):
            tmp = 2*np.pi*F*(x*np.cos(W)+y*np.sin(W))+P
            comp1 = complex(0,tmp)
            comp2 = complex(0,P)
            G[size + x ][size + y ] = k * np.exp(-np.pi*(S**2)*(x*x+y*y)) * (np.exp(comp1) - np.exp(-np.pi*((F/S) ** 2) + comp2))
    GABOUT = scipy.signal.convolve2d(I, G, 'same')
    GABOUT[np.imag(GABOUT) != 0] = np.real(GABOUT[np.imag(GABOUT) != 0])
    GABOUT[GABOUT < 0] = 0
    GABOUT = GABOUT.astype(np.double)
    return GABOUT
```
